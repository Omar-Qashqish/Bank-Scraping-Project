```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
from urllib.parse import urljoin
```

```python
url = "https://hbtf.com/ar"
base_url = "https://hbtf.com"
headers = {"User-Agent": "Mozilla/5.0"}
```

```python
folder_name = "Housing_Bank_Images"
os.makedirs(folder_name, exist_ok=True)

response = requests.get(url, headers=headers)
response.raise_for_status()
soup = BeautifulSoup(response.text, "html.parser")
```

```python
data = []

# استخراج كل الصور من السلايدر
slides = soup.select(".house_banner_sec_img")

for slide in slides:
    img_tags = slide.find_all("img")
    for img_tag in img_tags:
        img_src = img_tag.get("src", "").strip()
        if img_src:
            # نحاول نطلع الرابط الأب (Parent <a>) للصورة
            parent_link_tag = img_tag.find_parent("a", href=True)
            article_link = urljoin(base_url, parent_link_tag["href"]) if parent_link_tag else url

            # تجهيز رابط الصورة
            full_img_url = urljoin(base_url, img_src)
            image_name = img_src.split("/")[-1].split("?")[0]
            img_path = os.path.join(folder_name, image_name)

            try:
                img_resp = requests.get(full_img_url)
                img_resp.raise_for_status()
                with open(img_path, "wb") as f:
                    f.write(img_resp.content)
                print(f"✔️ تم تحميل الصورة: {image_name}")

                data.append({
                    "Bank_Name": "Housing Bank",
                    "Image_Name": image_name,
                    "Image_Link": full_img_url,
                    "Article_Link": article_link
                })
            except Exception as e:
                print(f"❌ خطأ تحميل الصورة: {full_img_url}, السبب: {e}")
```

```python
df = pd.DataFrame(data)
df
```
