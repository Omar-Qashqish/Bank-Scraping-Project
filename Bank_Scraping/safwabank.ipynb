```python
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin
```

```python
bank_name = "Safwa Bank"
base_url = "https://www.safwabank.com"
page_url = base_url + "/ar/"
output_folder = "safwabank_images"
os.makedirs(output_folder, exist_ok=True)
```

```python
url = "https://www.safwabank.com/ar/"

bank_name = "Safwa Bank"

output_folder = "safwa_images"
os.makedirs(output_folder, exist_ok=True)

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}


response = requests.get(url, headers=headers)
response.raise_for_status()
soup = BeautifulSoup(response.text, "html.parser")
```

```python
records = []

# دالة لتحميل الصور
def download_image(img_url):
    img_name = os.path.basename(img_url.split("?")[0])
    img_path = os.path.join(output_folder, img_name)

    # إذا كانت الصورة محمّلة مسبقًا، تجاهلها
    if os.path.exists(img_path):
        print(f"⚠️ الصورة موجودة مسبقًا: {img_name}")
        return img_name, img_url

    try:
        r = requests.get(img_url, timeout=10)
        r.raise_for_status()
        with open(img_path, "wb") as f:
            f.write(r.content)
        print(f"✅ تم تحميل الصورة: {img_name}")
        return img_name, img_url
    except Exception as e:
        print(f"❌ فشل تحميل {img_url}: {e}")
        return None, img_url
```

```python
# 1. صور السلايدر
for img in soup.select("rs-slide img, rs-layer img"):
    src = img.get("src") or img.get("data-lazyload") or img.get("data-src-rs-ref")
    if src:
        if src.startswith("//"):
            src = "https:" + src
        img_name, img_url = download_image(src)
        records.append({
            "Bank": bank_name,
            "Image_Name": img_name,
            "Image_Link": img_url,
            "Source": "Slider"
        })
```

```python
# 2. صور الأخبار
for img in soup.select("div.news-element img"):
    src = img.get("src")
    if src:
        img_name, img_url = download_image(src)
        records.append({
            "Bank": bank_name,
            "Image_Name": img_name,
            "Image_Link": img_url,
            "Source": "News"
        })
```

```python
# 3. صور قسم "صفوة موبايل"
for img in soup.select("div.vc_single_image-wrapper img"):
    src = img.get("src")
    if src:
        img_name, img_url = download_image(src)
        records.append({
            "Bank": bank_name,
            "Image_Name": img_name,
            "Image_Link": img_url,
            "Source": "Mobile"
        })
```
